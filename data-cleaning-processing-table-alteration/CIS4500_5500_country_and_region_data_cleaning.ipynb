Let's perform entity resolution. We'll bring in the cleaned World Bank data file so that we can ensure that the country names in the new table that we're introducing to the database are consistent with the existing data.

```python
world_bank_path = ''
world_bank = pd.read_csv(world_bank_path)
world_bank.columns
```

```python
cleaned_country_names = world_bank['Country Name'].unique()
cleaned_country_names
```

```python
new_raw_country_names = world_bank_country['Table Name'].unique()
new_raw_country_names
```

```python
# let's combine the country names into one set
countries = set(cleaned_country_names.tolist() + new_raw_country_names.tolist())
countries
```

```python
# we already have a list of the countries that had their entities resolved
# and can specifically target those country names
# for entity resolution

# the names below are:
# the names copied from the original data cleaning and processing notebook
# with additional names added at the end
# (it turned out there were still more entities to resolve,
# as there seem to be slight differences between World Bank files)

# key: name we're getting rid of
# value: name we're keeping
name_map = {
    'Antigua & Departments': 'Antigua and Barbuda',
    'Bahamas': 'Bahamas, The',
    'Bosnia Herzegovina': 'Bosnia and Herzegovina',
    'Brunei': 'Brunei Darussalam',
    'Congo (Brazzaville)': 'Congo, Rep.',
    'Congo (Democratic Republic)': 'Congo, Dem. Rep.',
    'Czechia': 'Czech Republic',
    'Egypt, Arab Rep.': 'Egypt',
    'Gambia': 'Gambia, The',
    'Iran, Islamic Rep.': 'Iran',
    'Ivory Coast': "Cote d'Ivoire",
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Lao PDR': 'Laos',
    'mexico': 'Mexico',
    'Micronesia': 'Micronesia, Fed. Sts.',
    'Sao Tome': 'Sao Tome and Principe',
    'Slovak Republic': 'Slovakia',
    'Korea, Rep.': 'South Korea',
    'Western Samoa': 'Samoa',
    'St Vincent & Grenadines': 'St. Vincent and the Grenadines',
    'Syrian Arab Republic': 'Syria',
    'Trinidad & Tobago': 'Trinidad and Tobago',
    'Turkiye': 'Turkey',
    'Venezuela, RB': 'Venezuela',
    'Viet Nam': 'Vietnam',
    'Yemen, Rep.': 'Yemen',
    "Côte d'Ivoire": "Cote d'Ivoire",
    'Curaçao': 'Curacao',
    'São Tomé and Principe': 'Sao Tome and Principe',
    'Türkiye': 'Turkey'
}

name_series = pd.Series(data=list(name_map.values()), index=list(name_map.keys()))
name_series
```

```python
# find indices of names to replace
purge_list = list(name_map.keys())
world_bank_country_idx = world_bank_country.index[world_bank_country["Table Name"].isin(purge_list)]

# using those indices, extract names that need to be updated as a series
world_bank_country_problems = world_bank_country.loc[world_bank_country_idx, "Table Name"]

# replace names
world_bank_country_fixed = name_series.loc[world_bank_country_problems.values].values

# update dataframe
world_bank_country.loc[world_bank_country_idx, "Table Name"] = world_bank_country_fixed
```

```python
# check if the names are fixed
set(world_bank['Country Name'].unique().tolist() + world_bank_country['Table Name'].unique().tolist())
```

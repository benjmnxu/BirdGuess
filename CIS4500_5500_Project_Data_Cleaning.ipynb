## Entity Resolution

The common entities between the Xeno Canto bird dataset and the World Bank dataset are the countries.

As instructed by the Processing Tutorial:


> To perform entity resolution, we will:
1. Determine whether both datasets use the same names to refer to all countries
2. Edit the names in one dataset to match the other, if necessary



Following the Processing Tutorial, "Let's compile a list of all country names in both datasets, then inspect it for repetitions."

"First, we extract the unique names of countries from both datasets."

```python
xeno_canto_bird_countries = xeno_canto_bird['country'].unique()
world_bank_countries = world_bank['Country Name'].unique()
```

"Now, we combine these into one set."

```python
countries = set(xeno_canto_bird_countries.tolist() + world_bank_countries.tolist())
```

"Let's inspect the contents of the set. The output is automatically sorted."

Let's resolve the inconsistent names.

```python
# key: name we're getting rid of
# value: name we're keeping
name_map = {
    'Antigua & Departments': 'Antigua and Barbuda',
    'Bahamas': 'Bahamas, The',
    'Bosnia Herzegovina': 'Bosnia and Herzegovina',
    'Brunei': 'Brunei Darussalam',
    'Congo (Brazzaville)': 'Congo, Rep.',
    'Congo (Democratic Republic)': 'Congo, Dem. Rep.',
    'Czechia': 'Czech Republic',
    'Egypt, Arab Rep.': 'Egypt',
    'Gambia': 'Gambia, The',
    'Iran, Islamic Rep.': 'Iran',
    'Ivory Coast': "Cote d'Ivoire",
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Lao PDR': 'Laos',
    'mexico': 'Mexico',
    'Micronesia': 'Micronesia, Fed. Sts.',
    'Sao Tome': 'Sao Tome and Principe',
    'Slovak Republic': 'Slovakia',
    'Korea, Rep.': 'South Korea',
    'Western Samoa': 'Samoa',
    'St Vincent & Grenadines': 'St. Vincent and the Grenadines',
    'Syrian Arab Republic': 'Syria',
    'Trinidad & Tobago': 'Trinidad and Tobago',
    'Turkiye': 'Turkey',
    'Venezuela, RB': 'Venezuela',
    'Viet Nam': 'Vietnam',
    'Yemen, Rep.': 'Yemen',
}
name_series = pd.Series(data=list(name_map.values()), index=list(name_map.keys()))
name_series
```

"Next, we find the indices of all usages of the names we're purging in both datasets"


```python
purge_list = list(name_map.keys())
xeno_canto_bird_idx = xeno_canto_bird.index[xeno_canto_bird["country"].isin(purge_list)]
world_bank_idx = world_bank.index[world_bank["Country Name"].isin(purge_list)]
```

"Using these indices, we extract the names that we need to update as `Series`"

```python
xeno_canto_bird_problems = xeno_canto_bird.loc[xeno_canto_bird_idx, "country"]
world_bank_problems = world_bank.loc[world_bank_idx, "Country Name"]
```

"Next, we replace the problematic names in these `Series` with the appropriate counterparts by indexing `name_series` with them.

*The operations in the cell below replace the names correctly because `name_series` maps each name we wanted to purge to the we wanted to replace it with.*"

```python
xeno_canto_bird_fixed = name_series.loc[xeno_canto_bird_problems.values].values
world_bank_fixed = name_series.loc[world_bank_problems.values].values
```

"Finally, we update the dataframes with the fixed names."

```python
xeno_canto_bird.loc[xeno_canto_bird_idx, "country"] = xeno_canto_bird_fixed
world_bank.loc[world_bank_idx, "Country Name"] = world_bank_fixed
```

Let's check if the names are fixed

```python
set(xeno_canto_bird['country'].unique().tolist() + world_bank['Country Name'].unique().tolist())
```
